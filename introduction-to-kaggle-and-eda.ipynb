## *Import Libraries*

Before working with any dataset, we need to import some important Python libraries. These libraries help us with data handling, visualization, and analysis:

- **pandas**: to load and manipulate data.
- **numpy**: for numerical operations.
- **matplotlib** and **seaborn**: for data visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's begin by loading our dataset using the Pandas library. This dataset contains employment and wage statistics from Pakistan between 2006 and 2023.

```python

# Load the CSV file
df = pd.read_csv("/kaggle/input/employment-and-wages-in-pakistan-20062023/employment_and_wages.csv")

# Display first 5 rows
df.head()
```

### *Understand the Structure of the Dataset*

Before we analyze the data, it's important to understand its structure.
The `.info()` method shows us:

- Number of rows and columns
- Column names and data types
- Count of non-null values


```python
df.info()
```

### *Summary Statistics*

We’ll use the `.describe()` method to view basic statistics such as:

- Mean, min, and max values
- Percentile ranges
- Standard deviation

This gives us an idea of the data’s range and distribution.


```python
df.describe()
```

###  *Check for Missing Values*

Missing values can affect our analysis.  
We’ll check for them using `.isnull().sum()` to identify any empty cells in each column.


```python
df.isnull().sum()
```

### *Rename Columns*

To make our code cleaner and easier to read, we’ll rename some of the longer column names.

For example:
- `Total_Employment` → `Total_Emp`
- `Total_Wages_Salaries` → `Total_Wages`


```python
df.rename(columns={
    'Total_Employment': 'Total_Emp',
    'Production_Workers_Employment': 'Prod_Workers_Emp',
    'Office_Staff_Employment': 'Office_Staff_Emp',
    'Total_Wages_Salaries': 'Total_Wages',
    'Production_Workers_Wages_Salaries': 'Prod_Workers_Wages',
    'Office_Staff_Wages_Salaries': 'Office_Staff_Wages'
}, inplace=True)

# Confirm column names
df.columns
```

### *Visualize Total Employment Over Time*

Now let’s create a line chart to observe how employment has changed over time.

We’ll use:
- `Year_Month` on the x-axis  
- `Total_Emp` on the y-axis

This will give us a clear view of the trend over time.


```python
import matplotlib.pyplot as plt

# Convert Year_Month to datetime
df['Year_Month'] = pd.to_datetime(df['Year_Month'], errors='coerce')

# Line plot
plt.figure(figsize=(10,5))
plt.plot(df['Year_Month'], df['Total_Emp'],marker = 'o')
plt.title('Total Employment Over Time')
plt.xlabel('Year')
plt.ylabel('Total Employment')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### *Compare Wages Over Time*

Let’s compare the wage trends of production workers and office staff.

We’ll use a line chart for both.
- `Prod_Workers_Wages`
- `Office_Staff_Wages`

This will help us see how their earnings varied across years.


```python
plt.figure(figsize=(10,5))
plt.plot(df['Year_Month'], df['Prod_Workers_Wages'], label='Production Workers', marker='o')
plt.plot(df['Year_Month'], df['Office_Staff_Wages'], label='Office Staff', marker='s')
plt.title('Wages Trend: Production Workers vs Office Staff')
plt.xlabel('Year')
plt.ylabel('Wages (Normalized)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  *Correlation Heatmap*

Let’s generate a correlation heatmap to find out how different numerical columns are related.

For example:
- Does higher employment mean higher wages?
- Are office staff wages closely related to total wages?

We’ll use Seaborn for visualization.


```python
import seaborn as sns
import numpy as np

# Only numerical columns
numeric_df = df.select_dtypes(include=[np.number])

# Correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
